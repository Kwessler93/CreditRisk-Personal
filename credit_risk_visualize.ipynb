```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Load your dataset 
credit_risk_df = pd.read_csv("credit_risk_dataset.csv")
credit_risk_df.head()
```

```python
# Calculate default rate by loan intent
default_rate_by_intent = credit_risk_df.groupby('loan_intent')['loan_status'].mean().sort_values()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
default_rate_by_intent.plot(kind='bar', color='skyblue')
plt.title('Loan Default Rate by Loan Intent')
plt.xlabel('Loan Intent')
plt.ylabel('Default Rate')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Create a box plot for person_age
plt.figure(figsize=(8, 5))
plt.boxplot(credit_risk_df['person_age'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of Applicant Age')
plt.xlabel('Age')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Most applicants are between 20 and 40 years old, with a peak around the mid-to-late 20s.

IQR bounds:

Lower bound: 12.5

Upper bound: 40.5

Outliers detected: 1,494 applicants have an age above 40.5, which are considered statistical outliers here.

The maximum age is 144, which could be a data entry error or an actual rare case.

```python
# Drop NaN values from employment length
emp_length_clean = credit_risk_df['person_emp_length'].dropna()

# Create a box plot for employment length
plt.figure(figsize=(8, 5))
plt.boxplot(emp_length_clean, vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Employment Length')
plt.xlabel('Years of Employment')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Median employment length: ~4 years.

Most values lie between 2 and 7 years.

The maximum value is 123 years, which is highly unrealistic and likely a data entry error.

IQR method bounds:

Lower bound: –5.5 (not meaningful since employment can't be negative)

Upper bound: 14.5

Outliers detected: 853 entries have values above 14.5 years.
